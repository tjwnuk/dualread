```python
import ebooklib
from ebooklib import epub
from ebooklib.utils import debug
from bs4 import BeautifulSoup
```

```python
book_pl = epub.read_epub("books/pl/Projekt Hail Mary - Andy Weir.epub")

def clean(html):
    soup = BeautifulSoup(html, 'html.parser')
    text = soup.get_text()

    # convert to full unicode
    text = text.encode('utf-8').decode('utf-8')

    return text

items = []
for item in book_pl.get_items_of_type(ebooklib.ITEM_DOCUMENT):
    items.append(item)
    
item = items[11]

content = item.get_body_content()
content = clean(content)

print(content)
```
